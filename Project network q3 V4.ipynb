```python
### K-means
# Lecture 6a & b

# K-Means will be used to analyse clusters, expectation is to find/create 3 clusters. 
#One left of the major road, one on the major and one on the right of this.
#Import k-means algorithm
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Extract node positions from the graph
node_positions = {node: (data['x'], data['y']) for node, data in G.nodes(data=True)}
node_coords = np.array(list(node_positions.values()))
node_ids = list(node_positions.keys())

# Apply KMeans with 3 clusters
kmeans = KMeans(n_clusters=2, random_state=0).fit(node_coords)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
inertia = kmeans.inertia_  # total distance to centroids

# Assign colors based on cluster label
colors = ['red', 'blue', 'green', 'orange', 'purple']
node_colors = [colors[label] for label in labels]

# Map nodes to cluster labels
node_cluster_labels = {node_ids[i]: labels[i] for i in range(len(node_ids))}

# Plot the clustered graph
fig, ax = ox.plot_graph(
    G,
    node_color=node_colors,
    node_size=30,
    edge_color='gray',
    show=False,
    close=False
)

# Plot centroids
for c in centroids:
    ax.plot(c[0], c[1], marker='X', markersize=12, color='black')

# Add cluster label text to each node
for node, (x, y) in node_positions.items():
    cluster_label = node_cluster_labels[node]
    ax.text(x, y, str(cluster_label), fontsize=8, color='white', ha='center', va='center')

plt.title(f"K-means Clustering (3 Clusters)\nTotal Distance to Centroids: {inertia:.2f}")
plt.show()


#### calculate total distance Q from each node to its cluster center#############
import networkx as nx
from shapely.geometry import Point
from scipy.spatial import KDTree

# Step 1: Build KDTree of node positions
positions_array = np.array(list(node_positions.values()))
node_ids_array = np.array(node_ids)  # For easier indexing
tree = KDTree(positions_array)

# Step 2: Find nearest node for each centroid
_, idxs = tree.query(centroids)
centroid_nodes = node_ids_array[idxs]  # safer than node_ids[i]


# Step 3: Calculate total network distance
total_network_distance = 0
node_to_centroid_dist = {}

for i, node in enumerate(node_ids):
    cluster = labels[i]
    centroid_node = centroid_nodes[cluster]
    try:
        # Use Dijkstra to find shortest path length
        distance = nx.shortest_path_length(G, source=node, target=centroid_node, weight='length')
        total_network_distance += distance
        node_to_centroid_dist[node] = distance
    except nx.NetworkXNoPath:
        # Handle disconnected node case
        node_to_centroid_dist[node] = np.nan

# Step 4: Print result
print(f"✅ Total network distance from all nodes to their cluster centers: {total_network_distance:.2f} meters")

print("Centroid nodes:", centroid_nodes)
```
